```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
from shapely.geometry import shape
from shapely.ops import unary_union
from matplotlib.colors import LinearSegmentedColormap

# Custom color map for forest
colors = [(0, 0.5, 0), (1, 1, 1)]
cmap = LinearSegmentedColormap.from_list('white_to_dark_green', colors, N=256)

# Load land cover data
land_cover_gdf_G = gpd.read_file('geonetwork_landcover_gha_gc_adg/gha_gc_adg.shp')
land_cover_gdf_C = gpd.read_file('geonetwork_landcover_civ_gc_adg/civ_gc_adg.shp')
land_cover_gdf_B = gpd.read_file('geonetwork_landcover_bfa_gc_adg/bfa_gc_adg.shp')
land_cover_gdf_T = gpd.read_file('geonetwork_landcover_tgo_gc_adg/tgo_gc_adg.shp')

# Filter for forest grid codes
forest = [110, 120, 60, 40, 41]
for df in [land_cover_gdf_G, land_cover_gdf_C, land_cover_gdf_B, land_cover_gdf_T]:
    df.dropna(subset=['GRIDCODE'], inplace=True)

land_cover_gdf_G = land_cover_gdf_G[land_cover_gdf_G["GRIDCODE"].isin(forest)].copy()
land_cover_gdf_G["Country"] = "Ghana"
print(land_cover_gdf_G["GRIDCODE"].unique())

land_cover_gdf_C = land_cover_gdf_C[land_cover_gdf_C["GRIDCODE"].isin(forest)].copy()
land_cover_gdf_C["Country"] = "CIV"
print(land_cover_gdf_C["GRIDCODE"].unique())

land_cover_gdf_B = land_cover_gdf_B[land_cover_gdf_B["GRIDCODE"].isin(forest)].copy()
land_cover_gdf_B["Country"] = "BFA"
print(land_cover_gdf_B["GRIDCODE"].unique())

land_cover_gdf_T = land_cover_gdf_T[land_cover_gdf_T["GRIDCODE"].isin(forest)].copy()
land_cover_gdf_T["Country"] = "Togo"
print(land_cover_gdf_T["GRIDCODE"].unique())

# Load country borders
ghana = gpd.read_file('Ghana.geojson')
print(ghana.columns)
print(ghana["NAME_1"].unique())

civ = gpd.read_file('CIV.geojson')
print(civ.columns)
print(civ["NAME_1"].unique())

bfa = gpd.read_file('BFA.geojson')
print(bfa.columns)
print(bfa["NAME_1"].unique())

tgo = gpd.read_file('TGO.geojson')
print(tgo.columns)
print(tgo["NAME_1"].unique())

# Ensure all geometries are valid
for gdf in [ghana, civ, bfa, tgo]:
    gdf = gdf[gdf.geometry.is_valid]

# Reproject to UTM zone 30N (meters-based)
projected_crs = 'EPSG:32630'
ghana = ghana.to_crs(projected_crs)
civ = civ.to_crs(projected_crs)
bfa = bfa.to_crs(projected_crs)
tgo = tgo.to_crs(projected_crs)

land_cover_gdf_G = land_cover_gdf_G.to_crs(projected_crs)
land_cover_gdf_C = land_cover_gdf_C.to_crs(projected_crs)
land_cover_gdf_B = land_cover_gdf_B.to_crs(projected_crs)
land_cover_gdf_T = land_cover_gdf_T.to_crs(projected_crs)

# Create 20 km buffer around Ghana
ghana_buffer = ghana.buffer(20000)

# Clip countries to that buffer
civ_clipped = gpd.clip(civ, ghana_buffer)
bfa_clipped = gpd.clip(bfa, ghana_buffer)
tgo_clipped = gpd.clip(tgo, ghana_buffer)

#define the full visible extent
visible_extent = gpd.GeoSeries(
    pd.concat([
        ghana.geometry,
        civ_clipped.geometry,
        bfa_clipped.geometry,
        tgo_clipped.geometry
    ], ignore_index=True)
).geometry.union_all().buffer(10000)

minx, miny, maxx, maxy = visible_extent.bounds
buffer = 20000  # 20 km inward padding

ax.set_xlim(minx + buffer, maxx - buffer)
ax.set_ylim(miny + buffer, maxy - buffer)


# Clip land cover to that buffer
land_cover_gdf_C = gpd.clip(land_cover_gdf_C, ghana_buffer)
land_cover_gdf_B = gpd.clip(land_cover_gdf_B, ghana_buffer)
land_cover_gdf_T = gpd.clip(land_cover_gdf_T, ghana_buffer)

# Combine land cover for plotting
combined_gdf = pd.concat([
    land_cover_gdf_G,
    land_cover_gdf_C,
    land_cover_gdf_B,
    land_cover_gdf_T
], ignore_index=True)

# Select regions of interest
Ashanti_region = ghana[ghana['NAME_1'] == 'Ashanti']
Bono_region = ghana[ghana['NAME_1'] == 'Bono'] #Brong_Ahafo region split up into 3 regions in 2019
GreaterAccra_region = ghana[ghana['NAME_1'] == 'GreaterAccra']
UpperEast_region = ghana[ghana['NAME_1'] == 'UpperEast']
```

```python
#create the WKT file from a csv containing site name, lat, long, and region
import pandas as pd

# Load the input CSV
df = pd.read_csv("Sampling_site_locations.csv", sep="\t")  # <- update filename if needed
print(df.columns.tolist())

# Create WKT from longitude and latitude
df["WKT"] = df.apply(lambda row: f"POINT ({row['longitude']} {row['latitude']})", axis=1)

# Keep only the required columns in the right order
df = df[["WKT", "name", "region"]]

# Save as the final file for your mapping script
df.to_csv("MYMAPS_GhanaSamples.csv", index=False)
```

```python
import re
from shapely.wkt import loads as wkt_loads

# Reload MYMAPS_GhanaSamples.csv properly
points_data = pd.read_csv("MYMAPS_GhanaSamples.csv")

# Use shapely to parse WKT strings
geometry = points_data["WKT"].apply(wkt_loads)
points_geo = gpd.GeoDataFrame(points_data, geometry=geometry, crs="EPSG:4326")  # assumes WGS84 lat/lon

# Reproject to match the map's CRS
points_geo = points_geo.to_crs("EPSG:32630")
```

```python
# Load river shapefile
river_data = gpd.read_file("ne_10m_rivers_lake_centerlines/ne_10m_rivers_lake_centerlines.shp")

# Reproject to match Ghana (EPSG:32630)
river_data = river_data.to_crs(ghana.crs)

# Bounding box filter (in meters)
river_data = gpd.clip(river_data, visible_extent)
```

```python
# Load and reproject lakes
lake_data = gpd.read_file("ne_10m_lakes/ne_10m_lakes.shp")
lake_data = lake_data.to_crs(ghana.crs)

# Filter for Lake Volta by name
lake_volta = lake_data[lake_data['name'].str.contains("Volta", case=False, na=False)]
```

```python
import matplotlib as mpl
from matplotlib import pyplot as plt
import matplotlib.patches as patches
from matplotlib.lines import Line2D
mpl.rcParams['font.family'] = 'sans-serif'

fig, ax = plt.subplots(figsize=(8, 8))
river_data.plot(ax=ax, color='#89CFF0', linewidth=2, zorder=3)
lake_volta.plot(ax=ax, color='#ADD8E6', edgecolor='black', zorder=2)

# Forest land cover
combined_gdf.plot(ax=ax, cmap=cmap, legend=True, zorder=2)

# Country borders
ghana.boundary.plot(ax=ax, edgecolor='black', linewidth=0.5, zorder=1)
civ_clipped.boundary.plot(ax=ax, edgecolor='black', linewidth=0.5, zorder=1)
bfa_clipped.boundary.plot(ax=ax, edgecolor='black', linewidth=0.5, zorder=1)
tgo_clipped.boundary.plot(ax=ax, edgecolor='black', linewidth=0.5, zorder=1)

# Ghana regions of interest
Ashanti_region.boundary.plot(ax=ax, color='black', linewidth=2, zorder=4)
Bono_region.boundary.plot(ax=ax, color='black', linewidth=2, zorder=4)
GreaterAccra_region.boundary.plot(ax=ax, color='black', linewidth=2, zorder=4)
UpperEast_region.boundary.plot(ax=ax, color='black', linewidth=2, zorder=4)

# Border highlight (thick line)
for clipped_country in [civ_clipped, bfa_clipped, tgo_clipped]:
    border_overlay = gpd.overlay(ghana, clipped_country, how='intersection', keep_geom_type=False)

    # Explode any GeometryCollections into individual geometries
    exploded = border_overlay.explode(index_parts=False)

    # Filter to only line-like geometries that can be plotted as boundaries
    line_like = exploded[exploded.geom_type.isin(["LineString", "MultiLineString"])]

    # Plot the cleaned borders
    line_like.plot(ax=ax, color='#960018', linewidth=3, zorder=5)


# Sampling points
points_geo.plot(ax=ax, marker='o', color='#FFA000', markersize=100, edgecolor='black', zorder=6)


plt.title('')
plt.xlabel('')
plt.ylabel('')

#region labels
ax.text(8.0e5, 5.85e5, "Greater Accra", size=14, fontweight='bold')
ax.text(6.5e5, 7.80e5, "Ashanti", size=14, fontweight='bold')
ax.text(5.25e5, 8.20e5, "Bono", size=14, fontweight='bold')
ax.text(
    7.10e5, 11.45e5, "Upper East",
    size=14, fontweight='bold',
    bbox=dict(facecolor='white', alpha=0.6, edgecolor='none', boxstyle='round,pad=0.3')
)

#points_geo[points_geo["longitude"]!="8.2585607"].plot(ax=ax, marker='o', color='#FFA000', markersize=50, zorder=3, edgecolor='black')
#points_geo[points_geo["longitude"]=="8.2585607"].plot(ax=ax, marker='o', color='#4997D0', markersize=50, zorder=3, edgecolor='black')

# Remove axes and ticks
ax.set_yticklabels([])
ax.set_xticklabels([])
ax.tick_params(axis='y', size=0)
ax.tick_params(axis='x', size=0)
[ax.spines[loc].set_visible(False) for loc in ['top','right','left','bottom']]

# Zoom in by tightening the map view around the visible extent
minx, miny, maxx, maxy = visible_extent.bounds
buffer = 20000  # in meters (adjust as needed)
ax.set_xlim(minx + buffer, maxx - buffer)
ax.set_ylim(miny + buffer, maxy - buffer)

# Annotations
ax.text(5.15e5, 5.0e5, "Ghana", size=22)
ax.text(4.25e5, 7.5e5, "Côte d’Ivoire Border", size=12, rotation=80)
ax.text(8.9e5, 1.1e6, "Togo Border", size=12, rotation=290)
ax.text(4.5e5, 1.245e6, "Burkina Faso Border", size=12, rotation=0) 

# Custom legend entries
legend_elements = [
    Line2D([0], [0], color='black', lw=4, label='Regions Sampled'),
    Line2D([0], [0], color='#960018', lw=4, label='Border'),
    Line2D([0], [0], color='#89CFF0', lw=4, label='River/Lake'),
    Line2D([0], [0], marker='o', color='w', label='Sampling Site',
           markerfacecolor='#FFA000', markeredgecolor='black', markersize=10),
]

# Add the legend
ax.legend(handles=legend_elements, loc='lower right', bbox_to_anchor=(1.1, -0.14), frameon=False, fontsize=12)

# Add a fake colorbar for forest gradient
from matplotlib.colorbar import ColorbarBase
from matplotlib.colors import Normalize
from mpl_toolkits.axes_grid1.inset_locator import inset_axes

# Create inset axis for the colorbar (below the legend)
cax = inset_axes(ax, width="30%", height="2%", loc='lower right',
                 bbox_to_anchor=(0.04, -0.2, 1, 1), bbox_transform=ax.transAxes, borderpad=1)

# Add the horizontal colorbar without ticks
cb = ColorbarBase(cax, cmap=cmap, norm=Normalize(vmin=0, vmax=1), orientation='horizontal')
cb.set_ticks([])

# Lower the label for spacing
cb.set_label('Forest cover', fontsize=12, labelpad=8)

# Add "More" and "Less" labels beside the colorbar ends
cax.text(0, 1.2, "More", va='bottom', ha='left', fontsize=10, transform=cax.transAxes)
cax.text(1, 1.2, "Less", va='bottom', ha='right', fontsize=10, transform=cax.transAxes)

# Save
plt.savefig("Map_Ghana_Final.png", bbox_inches='tight', dpi=1000)
plt.savefig("Map_Ghana_Final.svg", bbox_inches='tight')
```
